```python
import numpy as np
import matplotlib.pyplot as plt
import random
plt.rcParams['figure.figsize'] = [20, 10]
```

```python
random.seed(10)
np.random.seed(10)
```

```python
citiesnp = np.genfromtxt('cities.csv', delimiter=',', skip_header = 1) # load as numpy array
```

```python
plt.scatter(citiesnp[:,1], citiesnp[:,2], s=0.05)
```

## K-Means

```python
points = np.delete(citiesnp,0,1)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=5)
y_km = kmeans.fit_predict(points)
```

```python
y_km
```

```python
plt.scatter(points[y_km ==0,0], points[y_km == 0,1], s=0.1)
plt.scatter(points[y_km ==1,0], points[y_km == 1,1], s=0.1)
plt.scatter(points[y_km ==2,0], points[y_km == 2,1], s=0.1)
plt.scatter(points[y_km ==3,0], points[y_km == 3,1], s=0.1)
plt.scatter(points[y_km ==4,0], points[y_km == 4,1], s=0.1)
plt.scatter(points[y_km ==5,0], points[y_km == 5,1], s=0.1)
plt.scatter(points[y_km ==6,0], points[y_km == 6,1], s=0.1)
plt.scatter(points[y_km ==7,0], points[y_km == 7,1], s=0.1)
plt.scatter(points[y_km ==8,0], points[y_km == 8,1], s=0.1)
plt.scatter(points[y_km ==9,0], points[y_km == 9,1], s=0.1)
```
